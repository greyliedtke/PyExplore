```python
# indonesia weather data
import pandas as pd
df = pd.read_csv('datasets/climate_data.csv')
```

```python
import plotly.express as px
df = df.query('station_id == 96001')
fig = px.scatter(df, x='date', y='Tavg')
fig.show()
```

```python
# summarize to max / min of year
df['year'] = pd.DatetimeIndex(df['date']).year
# yearly_avgs = 
```

```python
# year summaries
years = pd.unique(df['year'])
y_avgs, y_mins, y_maxs = [], [], []
for y in years:
    ydf = df.query(f'year == {y}')
    y_avgs.append(ydf['Tavg'].mean())
    y_mins.append(ydf['Tn'].mean())
    y_maxs.append(ydf['Tx'].mean())
year_sum = pd.DataFrame({'year': years, 'Tavg': y_avgs, 'Tmax': y_maxs, 'Tmin': y_mins})
```

```python
import plotly.express as px
fig = px.line(year_sum, x='year', y=['Tavg', 'Tmax', 'Tmin'])
fig.show()
```

```python
# using year data from one section to predict next year...
import pandas as pd
df = pd.read_csv('datasets/climate_data.csv')
df = df.dropna(subset=['Tavg', 'Tx', 'ss'])
df['date'] = pd.DatetimeIndex(df['date'])
df['days_from_start'] = (df.date - df.date[0]).dt.days
ss = df[df.station_id == 96001]
ss2 = df[df.station_id == 96109]
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()  # create object for the class
x_data = ss['Tavg'].values.reshape(-1,1)
y_data = ss['ss'].values.reshape(-1,1)
lr.fit(x_data, y_data)  # perform linear regression

ss2['pred_sun'] = lr.predict(ss2['Tavg'].values.reshape(-1,1))  # make predictions

# coef = linear_regressor.coef_
# print(f"Coefficients: {coef}")

    # intercept = linear_regressor.intercept_
    # print(f"Intercept: {intercept}")

    # score = r2_score(y_data, y_pred)
    # print(f"Score: {score}")
```
