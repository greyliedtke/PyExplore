```python
# scrape from all recipes
from bs4 import BeautifulSoup
import requests

def word_descramble(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    return soup
```

```python

eispr = word_descramble('https://wordunscrambler.me/unscramble/eispr')

```

```python
res = eispr.find_all(class_="result")
for r in res:
    print(r)
    print(r.get_text())
```
