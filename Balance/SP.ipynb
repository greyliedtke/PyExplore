```python
# ServoPendulum creation...
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
g = 9.81    # gravity
l = .3      # meters


deg_range = np.linspace(0, np.pi)
deg_s2 = []

for r in deg_range:
    ay = math.sin(r)
    if ay ==0: alpha = 0
    else: alpha = 1*g*ay/l
    dt = math.degrees(alpha)
    deg_s2.append(alpha)

fig, ax = plt.subplots()
ax.plot(deg_range, deg_s2)
ax.grid()
```

```python
g = 9.81    # gravity
l = .3      # meters


deg_range = np.linspace(120,180)
deg_s2 = []

for r in deg_range:

    ax = -g*l*math.tan(math.radians(r))
    deg_s2.append(ax)

fig, ax = plt.subplots()
ax.plot(deg_range, deg_s2)
ax.grid()
```

```python
g = 9.81    # gravity
l = .3      # meters

deg_range = range(0,180)
deg_s2 = []

for d in deg_range:
    rad = math.radians(d)
    at = g*math.cos(rad)
    deg_s2.append(at)

fig, ax = plt.subplots()
ax.plot(deg_range, deg_s2)
ax.grid()
```
