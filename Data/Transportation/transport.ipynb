```python
import pandas as pd
df  = pd.read_csv('transport.csv')
df = df[0:48]   # getting data on states
cols = list(df.columns)

import matplotlib.pyplot as plt
# %matplotlib widget
```

```python
df.describe()
# df = df.interpolate()

df['State'] = df['State'].astype(str)
```

```python
def top_x_bar(df, y_col, Name, top_x=10):
    fig, ax = plt.subplots()
    df[y_col] = pd.to_numeric(df[y_col], errors='coerce')
    top_range = df.nlargest(top_x,y_col)
    ax.bar(top_range['State'], top_range[y_col])
    ax.set_title(f'Best cities for {Name}')
    plt.xticks(rotation=90)
    ax.set_xlabel('City')
    ax.set_ylabel('Score')
    plt.grid()
    # plt.show()
    fname = y_col.replace('/', '_')
    fname = fname.replace(' ', '_')
    fname = fname.replace('-', '_')
    fname = fname.replace(':', '_')
    save_name = f"Figs/{fname}.png"
    print(save_name)
    fig.savefig(save_name)
```
