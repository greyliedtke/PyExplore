```python
# importing libraries and data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import plotly.graph_objects as go

df = pd.read_csv('migrations.csv')
```

```python
# grouping examples
to_group = df.groupby(by='to')
python = to_group.get_group('Python')
golang = to_group.get_group('Golang')
to_counts = df['to'].value_counts()
from_counts = df['from'].value_counts()
comp_counts = df['company'].value_counts()
```

```python
# plot number of stars given
def bar_10(counts):
    counts = counts[0:10]
    fig, ax = plt.subplots()
    ax.bar(counts.index, counts.values)
    # ax.set_title(f'{counts}')
    ax.tick_params(rotation=45)
    plt.show()
for c in [to_counts, from_counts, comp_counts]:
    bar_10(c)
```

```python
# example sankey diagram from this...

# Define the input and output values and their labels
# each row represents from x to y
label = ["ZERO", "ONE", "TWO", "THREE", "FOUR", "FIVE"]
source = [0, 0, 1, 1, 0]
target = [2, 3, 4, 5, 4]
value = [8, 2, 2, 8, 4]

link = dict(source = source, target = target, value = value)
node = dict(label = label, pad=50, thickness=5)

data = go.Sankey(link = link, node=node)

fig = go.Figure(data)

fig.show()



```
