```python
import pandas as pd
df  = pd.read_csv('sp500.csv')
cols = df.columns

import matplotlib.pyplot as plt
# %matplotlib widget
```

### SP500 Analysis

A look into the S&P 500 to see the breakdown of year companies were added, industries, sub-industries, and top 5 states.


```python
# getting year added to find growing years in industry
df['date'] = pd.to_datetime(df['Date added'])
df['year'] = df['date'].dt.year
year_counts = df['year'].value_counts()
```

```python
# plot bar of new companies
fig, ax = plt.subplots()
ax.bar(year_counts.index, year_counts)
ax.set_title('SPX New Companies')
ax.set_xlabel('Year')
ax.set_ylabel('# Companies')
plt.grid()
plt.show()
```

```python
# pi chart of industries
industries = df['Industry'].value_counts()
plt.pie(industries, labels=industries.index, autopct='%1.1f%%')
plt.title('SPX Indsustry Breakdown')
plt.show()
```

```python
# breakdown of sub industries
for i in industries.index:
    sub_ind = df[df['Industry']== i]
    sub_ind = sub_ind['Sub-Industry'].value_counts()
    plt.pie(sub_ind, labels=sub_ind.index)
    plt.title(f'SPX {i} Breakdown')
    plt.show()
```

```python
# breakdown of states
df['state'] = df['Headquarters Location'].str.split(',').str[1].str[1:]
# plot bar of new companies
state_counts = df['state'].value_counts()
state_counts = state_counts[0:5]
fig, ax = plt.subplots()
ax.bar(state_counts.index, state_counts)
ax.set_title('SPX States')
ax.set_xlabel('State')
ax.set_ylabel('# Companies')
plt.grid()
plt.show()
```
