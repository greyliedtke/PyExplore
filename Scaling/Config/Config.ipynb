```python
import json
import pandas as pd
```

```python
# Load and parse JSON configuration file
with open('40kW_config.json', 'r') as json_file:
    config = json.load(json_file)

def device_config(device, config_file):
    # function to get all variables on system from xx
    dd = {}

    for instrumentation in config_file['instrumentation']:
        for sensor in config_file['instrumentation'][instrumentation]:
            sensor_info = config_file['instrumentation'][instrumentation][sensor]
            if sensor_info['system'] == device:
                dd[sensor] = sensor_info

    return dd

pi_ai = device_config('PI_AI', config)
pi_tc = device_config('PI_TC', config)


```

```python
# converting current instrumentation sheet to json
df = pd.read_excel('VariableConfig.xlsx')
df = df.sort_values(by='serialized_no')
grouped = df.groupby('description')
# Convert the groupby object to a dictionary
grouped_dict = {name: group.to_dict(orient='records') for name, group in grouped}

# Convert the dictionary to JSON
json_data = json.dumps(grouped_dict, indent=4)

# Print the JSON data
print(json_data)

```

```python
# building configuration references...
with open('Task07_Config.json', 'r') as json_file:
    config = json.load(json_file)

def device_config(device, sensor_type, config_file):
    # function to get all variables on system from xx
    dd = {}

    for instrument in config_file['instrumentation']:
        for sensor in config_file['instrumentation'][instrument]:
            if sensor['system'] == device:
                if sensor['type'] == sensor_type:
                    dd[sensor['influx_name']] = sensor

    return dd

```

```python
# NI ai and tc setup
ni_ai = device_config('NI', 'AI', config)
ni_tc = device_config('NI', 'TC', config)
plc_ai = device_config('PLC', 'AI', config)
plc_tc = device_config('PLC', 'TC', config)
```
