```python
# visualizing circle location
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()

# creating circle representative of ball
sphere_d = 2.5
circle = plt.Circle((0, sphere_d/8), radius=sphere_d/2, color='blue', fill=False)
ax.add_artist(circle)


ax.legend(['example', 'example'])
ax.set_title('title')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_xlim([-sphere_d, sphere_d])
ax.set_ylim([-sphere_d, sphere_d])

ax.set_aspect('equal')
plt.grid()
plt.show()
```
