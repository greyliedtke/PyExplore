```python
import numpy as np
import matplotlib.pyplot as plt

# Define the x values for the sine function
base_count = 5
res = 100
x = np.linspace(0, 2*np.pi*base_count, res*base_count)

# Calculate the y values for the sine function
y = np.sin(x)
plt.plot(x, y)

for bc in range(base_count):
    xn = x+(np.pi/2)*bc
    yn = y+1*bc

    xn = xn[0:-res*bc]
    yn = yn[0:-res*bc]
    plt.plot(xn, yn)


# Add a title and axis labels
plt.title('Sine Function')
plt.xlabel('x')
plt.ylabel('sin(x)')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the x values for the sine function
base_count = 5
res = 100
x = np.linspace(np.pi, 3*np.pi*base_count, res*base_count)

# Calculate the y values for the sine function
y = np.cos(x)
plt.plot(x, y)

for bc in range(base_count):
    xn = x+(np.pi)*bc
    yn = y+2*bc

    xn = xn[0:-res*(bc+1)]
    yn = yn[0:-res*(bc+1)]
    plt.plot(xn, yn)


# Add a title and axis labels
plt.title('CoSine Function')
plt.xlabel('x')
plt.ylabel('cos(x)')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Set the parameter t
t = np.linspace(0, 2*np.pi, 1000)

# Define the heart function using parametric equations
x = 16*np.sin(t)**3
y = 13*np.cos(t) - 5*np.cos(2*t) - 2*np.cos(3*t) - np.cos(4*t)

# Create a figure and axis object
fig, ax = plt.subplots()

# Plot the heart function
ax.plot(x, y)

# Set the aspect ratio to equal
ax.set_aspect('equal')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Set the radius of the circle
r = 1

# Set the parameter t
t = np.linspace(0, 2*np.pi, 1000)

# Define the x and y coordinates using parametric equations
x = r*np.cos(t)
y = r*np.sin(t)

# Create a figure and axis object
fig, ax = plt.subplots()

# Plot the circle
ax.plot(x, y)

# Add a small arc for the smile
smile = np.linspace(7*np.pi/6, 11*np.pi/6, 100)
x_smile = 0.6*r*np.cos(smile)
y_smile = 0.6*r*np.sin(smile) - r/3

ax.plot(x_smile, y_smile, linewidth=3)

# Add the eyes
eye_r = r/8
eye_x = [-r/3, r/3]
eye_y = [r/3, r/3]
for i in range(2):
    circle = plt.Circle((eye_x[i], eye_y[i]), eye_r, color='black')
    ax.add_artist(circle)

# Set the aspect ratio to equal
ax.set_aspect('equal')

# Show the plot
plt.show()
```
