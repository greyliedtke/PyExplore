```python
# function to get # of divisors
def divisors(number):
    divs = []
    for x in range(number):
        if number%(x+1):
            pass
        else:
            divs.append(x+1)
    return len(divs)
```

```python
# Loop through to 
ds = []
high_div = 0
high_divs = {}
numbers = range(1001)
for d in numbers:
    divs = divisors(d)
    ds.append(divs)
    if divs>high_div:
        high_div = divs
        high_divs[d] = divs
```

```python
# Plot
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
fig, ax = plt.subplots()

ax.scatter(numbers, ds)
ax.scatter(high_divs.keys(), high_divs.values(), color='orange')
ax.set_title('Divisors of Numbers')
ax.set_xlabel('Number')
ax.set_ylabel('Number of Divisors')
ax.grid()
ax.legend(['Numbers', 'Highly Composite Numbers'])
```

```python
# guasian distribution of data
# Calculate the mean and standard deviation of the data
mean = np.mean(ds)
std_dev = np.std(ds)

# Generate data points for the Gaussian distribution
# x = np.linspace(mean - 3*std_dev, mean + 3*std_dev, 1000)
x = numbers
y = (1/(std_dev*np.sqrt(2*np.pi))) * np.exp(-(x-mean)**2/(2*std_dev**2))

# Plot the Gaussian distribution
fig, ax = plt.subplots()

ax.plot(x, y, color='blue')

# Plot a histogram of the data for comparison
v_counts = pd.Series(ds).value_counts()
b = len(v_counts)
print(b)
ax.hist(ds, bins=b, density=True, color='gray', alpha=0.5)

# Set the plot title and axis labels
ax.set_title('Gausian Probability of composite numbers')
ax.set_xlabel('Divisors')
ax.set_ylabel('Probability Density')
ax.set_xlim([0,15])
ax.grid()

```

```python
# Plot the Gaussian distribution
fig, ax = plt.subplots()

ax.bar(v_counts.index, v_counts, color='blue')

# Set the plot title and axis labels
ax.set_title('Bar Plot of Divisors')
ax.set_xlabel('Divisors')
ax.set_ylabel('Number')
ax.set_xlim([0,15])
ax.grid()
```
