```python
# reading in data
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_excel('McMaster_21-23.xlsx')
df['line_cost'] = df['Quantity']*df['Unit Price']
```

```python
# group by order number
dfg = df.groupby('Purchase Order').agg({'line_cost': 'sum', 'Order Date': 'first'})
dfg = dfg.sort_values('line_cost', ascending=False)
print(dfg)
dfg.to_csv('mcm.csv')
```

```python
# plotting 
dfh = df.groupby('Purchase Order').agg({'line_cost': 'sum'})
plt.hist(dfh['line_cost'], bins=25)  # Specify the column and number of bins
plt.xlabel('Cost $')  # Set x-axis label
plt.ylabel('Frequency')  # Set y-axis label
plt.title('Histogram of McMaster Purchases')  # Set title
```

```python
fig, ax = plt.subplots()
ax.scatter(dfg['Order Date'], dfg['line_cost'])
ax.set_title('title')
ax.set_xlabel('Date')
ax.set_ylabel('$')
```
