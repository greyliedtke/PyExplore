```python
# imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
# cool wave

# Define the x values for the sine function
base_count = 2
res = 100
xn = np.linspace(0, np.pi*base_count, res*base_count)

# Calculate the y values for the sine function
y1 = 5*np.sin(xn)
y2 = 10*np.sin(xn)
y3 = 15*np.sin(xn)

c1 = "#b4bebf"
c2 = "#1f80ff"
# plt.axhspan(-20, 20, facecolor='#fffbad', alpha=1)

plt.fill_between(xn, y1, y2, where=(y2 > y1), interpolate=True, color=c1, alpha=1)
# plt.fill_between(xn, y1, 0, where=(y1 > 0), interpolate=True, color='green', alpha=0.5)
plt.fill_between(xn, y2, y3, where=(y3 > y2), interpolate=True, color=c2, alpha=1)

plt.fill_between(xn, y1, y2, where=(y2 < y1), interpolate=True, color=c1, alpha=1)
# plt.fill_between(xn, y1, 0, where=(y1 < 0), interpolate=True, color='green', alpha=0.5)
plt.fill_between(xn, y2, y3, where=(y3 < y2), interpolate=True, color=c2, alpha=1)

```

```python
# damping plot

# Define the x values for the sine function
base_count = 3
res = 100
xn = np.linspace(0, np.pi*base_count, res*base_count)

# Calculate the y values for the sine function
y1 = (base_count*np.pi-xn)*np.sin(xn)/(base_count*np.pi)
y2 = -y1

# plt.plot(xn, y1)
plt.fill_between(xn, y1, 0, where=(y1 > 0), interpolate=True, color=c1, alpha=1)
plt.fill_between(xn, 0, y1, where=(y1 < 0), interpolate=True, color=c1, alpha=1)
plt.fill_between(xn, y2, 0, where=(y2 > 0), interpolate=True, color=c2, alpha=1)
plt.fill_between(xn, 0, y2, where=(y2 < 0), interpolate=True, color=c2, alpha=1)
```

```python
# heart

# Set the parameter t
t = np.linspace(0, 2*np.pi, 1000)

# Define the heart function using parametric equations
x = 16*np.sin(t)**3
y = 13*np.cos(t) - 5*np.cos(2*t) - 2*np.cos(3*t) - np.cos(4*t)

# Create a figure and axis object
fig, ax = plt.subplots()

# Plot the heart function
ax.plot(x, y)

# Set the aspect ratio to equal
ax.set_aspect('equal')
```
